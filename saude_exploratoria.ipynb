# Análise Exploratória dos Dados de Saúde
Este notebook realiza uma análise exploratória dos dados de saúde mental já preparados, incluindo estatísticas descritivas, visualizações e filtragem de subgrupos relevantes.

## 1. Importar Bibliotecas Necessárias
Importe as bibliotecas para manipulação e visualização dos dados.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Carregar Dados de Saúde Preparados
Carregue o arquivo de dados já preparado para análise.

```python
# Carregar dados preparados
data = pd.read_csv('./data/prepared_data.csv')
data.head()
```

## 3. Visualizar Estrutura dos Dados
Exiba as primeiras linhas, tipos de dados e informações gerais do DataFrame.

```python
# Visualizar primeiras linhas, tipos e estatísticas
print('Primeiras linhas:')
display(data.head())
print('\nInfo:')
data.info()
print('\nEstatísticas descritivas:')
display(data.describe(include='all'))
```

## 4. Análise Exploratória: Estatísticas Descritivas
Calcule estatísticas como média, mediana, desvio padrão e valores ausentes para as principais variáveis.

```python
# Estatísticas descritivas principais
print('Média das variáveis numéricas:')
print(data.mean(numeric_only=True))
print('\nMediana das variáveis numéricas:')
print(data.median(numeric_only=True))
print('\nDesvio padrão das variáveis numéricas:')
print(data.std(numeric_only=True))
print('\nValores ausentes por coluna:')
print(data.isnull().sum())
```

## 5. Visualização de Distribuições de Variáveis
Utilize histogramas, boxplots e violinos para visualizar as distribuições das variáveis numéricas e categóricas.

```python
# Histogramas das variáveis numéricas
num_cols = data.select_dtypes(include='number').columns
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col].dropna(), kde=True)
    plt.title(f'Distribuição de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequência')
    plt.tight_layout()
    plt.show()

# Boxplots das variáveis numéricas
plt.figure(figsize=(10, 6))
sns.boxplot(data[num_cols])
plt.title('Boxplot das Variáveis Numéricas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 6. Análise de Correlação Entre Variáveis
Calcule e visualize a matriz de correlação entre variáveis numéricas usando heatmap.

```python
# Matriz de correlação entre variáveis numéricas
corr = data[num_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlação das Variáveis Numéricas')
plt.tight_layout()
plt.show()
```
